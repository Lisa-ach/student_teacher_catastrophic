```python
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
results_folder = "../run/results/2021-03-22-18-32-49"
```

```python
dfs = {f: pd.read_csv(os.path.join(results_folder, f"feature_{f}", "0", "ode_log.csv")) for f in np.linspace(0, 1, 20)}
```

```python
dfs[0].head
```

```python
fig = plt.figure(figsize=(16, 9))
for f, df in dfs.items():
    log_gen_1 = df["log_generalisation_error_1"]
    plt.plot(range(len(log_gen_1)), log_gen_1, label=f)
plt.legend()
fig.show()
```

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
path = "/Users/sebastianlee/Dropbox/Documents/Research/Projects/catastrophic/run/results/not_so_long_run/"
```

```python
dfs = {overlap: pd.read_csv(os.path.join(path, f"v_{overlap}", "0", "ode_log.csv")) for overlap in np.linspace(0, 1, 25)}
```

```python
color_map = cm.get_cmap("viridis")

fig = plt.figure(figsize=(15, 9))
for i, (overlap, df) in enumerate(dfs.items()):
    error_0 = np.array(df["log_generalisation_error_0"])
    plt.plot(range(len(error_0)), error_0, color=color_map(i / len(dfs)), linewidth=5, alpha=0.8)
plt.xlim(0, 7000)
plt.ylim(-3.5, -0.5)
plt.minorticks_on()
plt.grid(b=True, which='major', color='r', linestyle='-', alpha=0.2)
plt.grid(b=True, which='minor', color='gray', linestyle='--', alpha=0.2)
plt.tick_params(
                axis='both',         # changes apply to the x-axis
                which='both',      # both major and minor ticks are affected
                bottom=False,      # ticks along the bottom edge are off
                top=False,         # ticks along the top edge are off
                labelbottom=False, # labels along the bottom edge are off
                left=False,
                labelleft=False
            )
fig.show()
fig.savefig("no_convergence.pdf", dpi=100, bbox_inches='tight', pad_inches=0)
```
